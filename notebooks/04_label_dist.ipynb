```python
import pandas as pd 
import numpy as np
import os

import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
vehicle_names = ["SEB880", "SEB882", "SEB883", "SEB885", "SEB888", "SEB889"]
#vehicle_names = ["SEB880"]

PATH_TO_LOAD = "../Processed_data_new/04_Merged"

df = pd.DataFrame()

for vehicel in vehicle_names:
    df_curr = pd.read_csv(os.path.join(PATH_TO_LOAD, f"{vehicel}_merged.csv"))
    df = pd.concat([df, df_curr])

df= df[["domain", "Label", "session"]]
df = df[df.domain.notna()]
```

```python
print("Num labels: ", len(df))
print("Num sessions: ", len(df.session.unique()))
```

```python
driveMode_map = {
    "0.0": "Normal",
    "0": "Normal",
    "1.0": "Sport",
    "2.0": "SuperSport",
    "2": "SuperSport",
    "3.0": "Range",
    "3": "Range",
    "4.0": "Gravel-Offroad"
}
#df['Label'] = df['Label'].map(lambda x: driveMode_map[x] if x in driveMode_map else x)
```

```python
count = df.Label.value_counts()
df_sb = pd.DataFrame()
df_sb["domain"] = count.index.str.split('/').str[0]
df_sb["sub_domain"] = count.index.str.split('/').str[1]
df_sb["label"] = count.index.str.split('/').str[-1]
df_sb["label"] = df_sb["label"].map(lambda x: driveMode_map[x] if x in driveMode_map else x)
df_sb["count"] = np.log1p(count.values)
df_sb
```

```python
import plotly

fig = px.sunburst(
    df_sb,
    path=['domain', 'sub_domain', 'label'],  # Define the hierarchical path
    values='count',  # Define the values to be represented by the size of segments
    title='Sunburst Chart of Label Domains and Labels',
    width=1000,  
    height=1000
)

# Customize font size conditionally
segment_sizes = df_sb.groupby(['domain', 'label'])['count'].sum().reset_index()
segment_sizes['font_size'] = segment_sizes['count'].apply(lambda x: 16 if x >= 15 else 10)

fig.update_traces(
    textfont=dict(size=segment_sizes['font_size'])
)

fig.write_html('sunburst_chart.html', config={'staticPlot': False, 'responsive': True})
plotly.io.write_image(fig, '/Users/wannerje/Dropbox/Porsche/eth_report/Img/sunburst_chart.pdf', format='pdf')
```
